```python
###### pandas experimentation

sensor_data_frame = pd.DataFrame( columns=[ "locID", "locID_2", "PMid", "humidityID", "has_data", "sensor_type", "sensor_manufacturer", "sensor_value", "location_data", "humiditySensorName", "humdiditySensorValue" ] )
sensor_data_frame.set_index( "locID" )

sensor_data_frame
```

```python
sensor_data_frame['locID'] = location_id_list_freiburg

sensor_data_frame.head()
```

```python
original_Freigburg_loc_id_sensor_data = []
width_data_received_data_location_ids = []
no_data_received_data_location_ids = []
```

```python
# Filling real pandas data table with data… 

for loc_id in location_id_list_freiburg:
    print("\n ------- fetching loc id "+str( loc_id )+" -------")
    
    # fetch current data 
    current_sensor_json = json.loads( requests.get( "http://api.luftdaten.info/v1/sensor/"+str(loc_id)+"/" ).text )
    
    # if there is no data 
    if current_sensor_json == []:
        print("//// no data for id "+str( loc_id)+" ///////" )
        no_data_received_data_location_ids.append( loc_id )
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, 'has_data' ] = False
    
    # if there IS data 
    else  :
        width_data_received_data_location_ids.append( loc_id )
        original_Freigburg_loc_id_sensor_data.append( current_sensor_json )
        print("|| GOT DATA! for loc id "+str( loc_id)+"|||" )
        print( original_Freigburg_loc_id_sensor_data[-1] )
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, 'has_data' ] = True
        ## sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, 'location_data' ] = str( current_sensor_json )
        # --- fetch PM sensor data… 
        
        # eg sensor type 
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, 'sensor_type' ] = current_sensor_json[0]['sensor']['sensor_type']['name']
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, "sensor_manufacturer" ] = current_sensor_json[0]['sensor']['sensor_type']['manufacturer']
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, "locID_2" ] = current_sensor_json[0]['location']['id']
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, "PMid" ] = current_sensor_json[0]['sensor']['id']
        sensor_data_frame.loc[ sensor_data_frame['locID'] == loc_id, "humidityID" ] = current_sensor_json[0]['sensor']['id'] + 1
        # print("\n sensor data shape/length : "+str( len( current_sensor_json )))
        # print( current_sensor_json[0] )
        # print( current_sensor_json[0]['sensor']['sensor_type']['name'] )        
```

### now fill in the Humaidity data values, by fetching the humidity sensor data, using the computed sensor id ( based on the PM sensor id ) 



```python
for humiditySensorID in sensor_data_frame[ sensor_data_frame['has_data'] == True ]['humidityID']:
    #
    print("\n --------- fetching humiditySensorID |"+str( humiditySensorID )+"| - type : "+str( type( humiditySensorID )) )
    current_humidity_sensor_data = json.loads( requests.get( "http://api.luftdaten.info/v1/sensor/"+str(humiditySensorID)+"/" ).text )
    # data loading feedback
    print( "- len: "+str( len( current_humidity_sensor_data)) )
    print( current_humidity_sensor_data )
    
    # DEBUG 
    print( "\n ---- testing what to insert as humiditysensorname ")
    #
    auxillory_sensor_data_length = len( current_humidity_sensor_data ) 
    print("----- len : "+str( auxillory_sensor_data_length ) )
    
    if auxillory_sensor_data_length > 0:
        #
        print( current_humidity_sensor_data[-1] )
        print( "-------- sensor name : "+current_humidity_sensor_data[-1]['sensor']['sensor_type']['name'] )

        print( "\n ---- current humiditysensordata  ")
        # set relevant sensor data 
        sensor_data_frame.loc[ sensor_data_frame['humidityID'] == humiditySensorID, 'humiditySensorName' ] = current_humidity_sensor_data[-1]['sensor']['sensor_type']['name'];
        
        # find the humidity data
        curr_humidity_value = "no_data"
        # loop through sensor values and find humidity data 
        for sensor_data in current_humidity_sensor_data[-1]['sensordatavalues']:
            if sensor_data[ 'value_type' ] == "humidity":
                curr_humidity_value = sensor_data[ 'value' ]
        
        sensor_data_frame.loc[ sensor_data_frame['humidityID'] == humiditySensorID, 'humdiditySensorValue' ] = curr_humidity_value 
    
```
